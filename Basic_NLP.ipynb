## 5.1 similarité syntaxique 

Pour la similarité syntaxique entre des vecteurs, plusieurs distances sont possibles à savoir : distance, euclidienne, Cosine Jaccard, Levenshtein, Hamming…

l'exemple ci-dessous permet de calculer la similarité. semantique entre les documents en se basant sur une representation vectorielle en TFIDF avec la distance euclidienne. 

```python
from sklearn.metrics.pairwise import cosine_distances,euclidean_distances
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import Binarizer
import time

    
def tfidf_simil(corpus,index_of_doc):
   
    #tfidf_matrix = tfidf_matrix.toarray()
    #compute similarity for first sentence with rest of the sentences
    return (cosine_distances(corpus[index_of_doc:index_of_doc+1],corpus))

def ohv_simil(corpus,index_of_doc):
    vectorizer  = CountVectorizer()
    corpus=[ sent.lower() for sent in corpus]
    corpus = vectorizer.fit_transform(corpus)

    onehot = Binarizer()
    ohv = onehot.fit_transform(corpus.toarray())

    return (cosine_distances(ohv[index_of_doc:index_of_doc+1],ohv))

def bow_simil(corpus,index_of_doc):
    vectorizer = CountVectorizer(lowercase=True)
    corpus=[ sent.lower() for sent in corpus]
    X = vectorizer.fit_transform(corpus) #sparsy format
    bow = X.toarray()
    
    return (cosine_distances(bow[index_of_doc:index_of_doc+1],bow))
```

**Realiser la même chose en se basant sur les representations OHV et BOW?
**Comparer les performances des trois methodes en terme de temps d'execution et precision.  

```python
my_corpus = ['This is the first document.','This is the second second document.','And the third one.','Is this the first document?']

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(my_corpus)

start_time = time.time()
print(tfidf_simil(tfidf_matrix,0))
end_time = time.time()
print('tfidf executed in',end_time-start_time,'s')
```

```python
start_time = time.time()
print(ohv_simil(my_corpus,0))
end_time = time.time()
print('tfidf executed in',end_time-start_time,'s')
```

```python
start_time = time.time()
print(bow_simil(my_corpus,0))
end_time = time.time()
print('tfidf executed in',end_time-start_time,'s')
```

<b> Interprétation : TFIDF est plus précis mais BOW est plus rapide </b>

Il existe d'autre methodes(orientées caractères) pour le calcule de la similarité syntaxique entre des documents courts (des phrases):

* Longest Common Sequence (LCS)
* Set features

**Defnir la fonction SemanticDistanceDocs(doc1,doc2) qui permet de calculer la distance semantique totale entre deux documents texte

```python
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize

# définir une fonction pour calculer le sens d'un mot selon le contexte
def word_sem(word,context):
    sens=lesk(context, word)
    return sens

# fonction pour calculer la similarité entre deux docs
def SemanticDistanceDocs(doc1,doc2):
    
    # tokeniz the two docs
    contextDoc1 = word_tokenize(doc1)
    contextDoc2 = word_tokenize(doc2)
    
    # retourne les sens de tous les mots
    sensDoc1=[word_sem(word,contextDoc1) for word in contextDoc1]
    sensDoc2=[word_sem(word,contextDoc2) for word in contextDoc2]
    
    # Filtrer les None
    sensDoc1 = [ss for ss in sensDoc1 if ss is not None]
    sensDoc2 = [ss for ss in sensDoc2 if ss is not None]
    
    sim = []
    
    ## pour chaque sens du doc1 on calcule la similarité à l'aide de la fonction path_similarity
    for s1 in sensDoc1:
        best_score = []
        for s2 in sensDoc2:
            score = s1.path_similarity(s2)
            
            if score is not None:
                best_score.append(score)

        if(best_score):
            sim.append(max(best_score)) # on stocke le similarité du deux mot les plus similaire
            
        
    return sum(sim)/len(sim) # le simalirité entre les deux docs est la moyenne de tous les similarité


# la fonction SemanticDistanceDocs retourne deux score différent pour doc1,doc2 et doc2,doc1
# on va crée une fonction retourne une similarité symétrique
def SymSemanticDistanceDocs(doc1,doc2):
    return (SemanticDistanceDocs(doc1,doc2)+SemanticDistanceDocs(doc2,doc1))/2
```
